Write a Python class using NumPy that implements a **NAND logic gate** using a single perceptron. Your implementation must use **vectorized operations** (no explicit `for` loops in the activation function) to process an input matrix of shape `(N, 2)` and return the corresponding binary outputs.

```python
#code here
class PerceptronNAND:

  def __init__(self):
    self.weights = np.array([-2.0,-2.0])
    self.bias = 3.0

  def step_activation(self,x):
    return np.where(x>=0,1,0)

  def operate(self,inp):
    inp = np.asarray(inp)
    out = np.dot(inp,self.weights) + self.bias
    return self.step_activation(out)

if __name__ == "__main__":

  eg = np.array([[0,0],[1,0],[0,1],[1,1]])
  nand = PerceptronNAND()
  res = nand.operate(eg)

  for a,b in zip(eg,res):
    print(a[0],'nand',a[1],'=',b)
```

### Q3c - Too-Large Learning Rate

If you set eta = 0.6, what goes wrong on the b update? Show numerically. What does this tell you about learning rate selection?

> **Your answer:**
$$\frac{\partial C}{\partial b} = 4(b - 1)$$
$$b_{t+1} = b_t - \eta \frac{\partial C}{\partial b}$$
Substituting $\eta = 0.6$:
$$b_{t+1} = b_t - 0.6 \cdot 4(b_t - 1)$$$$b_{t+1} = b_t - 2.4(b_t - 1)$$
Distance from optimum b* = 1
$$b_{t+1} - 1 = b_t - 1 - 2.4(b_t - 1)$$$$b_{t+1} - 1 = -1.4(b_t - 1)$$
This shows that at each iteration, the error (distance from the optimum $b^*=1$) is multiplied by $-1.4$.

>For example, let's start with $b_0$ = 0 \:

>**Upper Bound for Stability**: For a quadratic cost function, gradient descent only converges if the learning rate satisfies $\eta < \frac{2}{L}$, where $L$ is the second derivative (or maximum eigenvalue of the Hessian matrix) along that dimension. ## Q4 - Sigmoid Neuron: Implementation and Visualisation

```python
class SigmoidNeuron:

    def __init__(self, weights, bias):
        self.weights = np.array(weights, dtype=float)
        self.bias = float(bias)

    def forward(self, x):
        # TODO: compute z = w.x + b, return sigmoid(z)
        z = np.dot(self.weights,x) +  self.bias
        sig_z = 1.0 / (1.0 + np.exp(-z))
        return sig_z

    def numerical_gradient(self, x, h=1e-5):
        # TODO: centered difference (f(x+h) - f(x-h)) / 2h
        grad = ( self.forward(x+h) - self.forward(x-h) ) / (2*h)
        grad_weights = grad*x
        grad_bias = grad*1
        return (grad_weights, grad_bias)
```

## Q5 - Gradient Descent on a Toy Regression Problem

A single-layer network (one input, one output, no activation) must learn y = 3x + 1.

```python
X_train = np.array([0.0, 1.0, 2.0, 3.0, 4.0])
y_train = np.array([1.0, 4.0, 7.0, 10.0, 13.0])
```

### Q5a - Analytical Gradients + Full-Batch GD

Derive dC/dw and dC/db analytically (show algebra). Implement full-batch GD with eta = 0.01 for 200 epochs. Plot the loss curve. Do w and b converge to 3 and 1?

**Derivation (show algebra):**

> **Your answer:**
>$$ C(w,b) = \frac{1}{2N} \sum_{i=1}^{N} (\hat{y}_i - y_i)^2, $$\
>$\hat{y}_i$ -> y with current (w,b) , $y_i$ -> True y value
>$$\frac{\partial C}{\partial w} = \frac{1}{N} \sum_{i=1}^{N} (\hat{y}_i - y_i) \cdot x_i$$
>$$\frac{\partial C}{\partial b} = \frac{1}{N} \sum_{i=1}^{N} (\hat{y}_i - y_i)$$


```python
w, b = 0.0, 0.0
eta = 0.01
losses = []
N = len(X_train)

for epoch in range(200):
    # TODO: predictions, loss, gradients, update
    y_pred = w*X_train + b

    dw = (1/N) * np.sum( (y_pred - y_train) * X_train)
    db = (1/N) * np.sum( y_pred - y_train )

    w -= eta*dw
    b -= eta*db

    loss = (1/(2*N)) * np.sum( (y_pred - y_train)**2 )
    losses.append(loss)

print(f'Final w = {w:.4f}, b = {b:.4f}')
plt.figure(figsize=(7, 4))
plt.plot(losses)
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Full-Batch GD (eta=0.01)')
plt.grid(True)
plt.show()
```

### Q5b - SGD with Mini-Batch Size = 1

Implement SGD with batch size 1. Train 200 epochs with eta = 0.01. Plot the loss curve. How does it differ from full-batch? Explain in terms of gradient quality and noise.

> **Your answer:**
>
>**Gradient Quality :**  full batch gradient descent computes exact gradient of the cost function across all training data, and the gradient vector points exactly in the direction of steepest descent for the global dataset.
The SGD calculates a heavily approximated gradient based on only one randomly selected data at a time. At any time the update given by any single data might differ significantly from the true global minimum.
>
>**Noise :** For full batch,Because the gradient quality is perfect, the optimization path is completely smooth, monotonic and deterministic. Loss curve drops smoothly towards zero without any fluctuation.
>For SDG,Because each single-sample update is imperfect, a significant amount of stochastic noise is introduced into the training trajectory. When looking closely at the loss curve, SGD exhibits noticeable fluctuations and oscillations.

```python
w, b = 0.0, 0.0
eta = 0.01
losses_sgd = []
N = len(X_train)


for epoch in range(200):
    # TODO: pick random example, compute gradient, update
    loss = 0
    indices = np.arange(N)
    np.random.shuffle(indices)

    for i in indices :
      y_true = y_train[i]
      x = X_train[i]

      y_pred = w * x + b

      loss += 0.5 * (y_pred - y_true) **2

      dw = (y_pred - y_true) * x
      db = (y_pred - y_true)

      w -= eta * dw
      b -= eta * db

    losses_sgd.append(loss)

print(f'Final w = {w:.4f}, b = {b:.4f}')
plt.figure(figsize=(7, 4))
plt.plot(losses_sgd)
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('SGD batch=1 (eta=0.01)')
plt.grid(True)
plt.show()
```

## Q7 - Geometric Interpretation of 1D Gradient Descent

Consider C(v) as a function of a single scalar v.

1. Give a precise **geometric interpretation** of the update rule v = v - alpha * dC/dv.
2. How do the slope and the learning rate alpha physically dictate movement along the curve?

> **Your answer:**
>
> 1. The Update rule can be interpreted as moving towards the minimum of the curve. It denotes the iterative process of walking down the hill towards the lowest point. It is taking a step of magnitude $|α*dc/dv|$ along steepest descent from the initial point v.
>
> 2. dc/dv (the slope) gives the direction of steepest ascent. So -dc/dv (-slope) gives the direction of steepest descent. The learning rate(α) determines the physical size of the step you take along v-axis.
> \
>\
>The update step is proportional to dc/dv indicating the step naturally scales with steepness of the curve. If v is far from minimum(valley) dc/dv will be large and steps are large. If v is close to minimum(valley) dc/dv will be small and so are the steps.
>\
>\
> Alpha controlls the physical dynamics of movement. Too small α gives tiny updates and takes large time to reach the bottom. Optimal α gives large step when slope is steep, and tiny step as it reaches the bottom. Too large α, gives very large step, so v physically leaps over the minimum to the opposite side landing higher up than it started. This causes divergence.

```python
v_vals = np.linspace(-3, 5, 300)
C_v = (v_vals - 2)**2 + 1

def c(v):
  return (v - 2)**2 + 1

def dc(v):
  return 2*(v - 2)

epoches = 10
v = -3
v_history = [v]
c_history = [c(v)]
eta = 0.1
# TODO: run gradient descent steps and annotate on the curve

for epoch in range(epoches):
  v = v - eta*dc(v)

  v_history.append(v)
  c_history.append(c(v))


plt.figure(figsize=(7, 4))
plt.plot(v_vals, C_v, label='C(v)')
plt.xlabel('v')
plt.ylabel('C(v)')
plt.title('1D Gradient Descent')
plt.legend()
plt.grid(True)


for i, (v, c) in enumerate(zip(v_history, c_history)):
    plt.annotate(f'v{i}',
                 xy=(v, c),
                 xytext=(10, -5),
                 textcoords='offset points',
                 fontsize=9,
                 color='darkred',
                 weight='bold')

plt.show()
```

> **Your answer:**

>1. When we remove all hidden layers, the deep neural network collapses into a foundational statistical model. When the 784 inputs are directly connected to 10 outputs and trained with SGD, it creates a Multinomial Logistic Regression model. Mathematically the network computes a simple linear transformation followed by a softmax activation function to turn scores to probabilities.
>$$P(y = k \mid \mathbf{x}) = \frac{e^{\mathbf{w}_k^T \mathbf{x} + b_k}}{\sum_{j=1}^{10} e^{\mathbf{w}_j^T \mathbf{x} + b_j}}$$
>2. The accuracy of this model is strictly limited because it can only learn linear decision boundaries.
> Each output neuron defines a hyperplane in the 784-dimensional input space and decides the class of an input based on which side of these hyperplane the data point falls. It cannot handle Non-Linear Relationships or Non-Linear logic problems like XOR.
>In an Image dataset, zero-hidden layer modal learns a blurry 'average template' for each digit. If the image is highly rotated, shifted or styled differently from average template, the linear combination fails to recognize it.
>3. This implies the absolute necessity of hidden layers pointing to 2 core principles:
>> **The Universal Approximation Theorm :** It states that a feed forward network need atleast one hiddden layer with a non-linear activation function(like sigmoid) to approximate any continuous function. Else we would be trapped in strict linearity.
>>
>> **Feature Representation Learning :** without hidden layers, the modal must make a decision using raw data(pixels).
>>With hidden layers, the network uses early layers to construct a hierarchy of features(from raw pixels -> edges -> shapes -> complex concepts ). Thus final layer still performs linear classification, but it does so on a linearly separable, highly optimized data space made by hidden layers.
