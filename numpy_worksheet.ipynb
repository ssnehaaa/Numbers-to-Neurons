### Exercise 15 — Vectorize a Function

**Question:** Convert the scalar function `maxx` below to work on two arrays.


```python
# Input — do not modify
def maxx(x, y):
    """Get the maximum of two items"""
    if x >= y:
        return x
    else:
        return y

a = np.array([5, 7, 9, 8, 6, 4, 5])
b = np.array([6, 3, 4, 8, 9, 7, 1])
```

```python
# Write your code here:
pair_max = np.vectorize(maxx, otypes=[float])
print(pair_max(a, b))
```

### Exercise 49 — Row-wise Unique Counts

**Question:** Compute the counts of unique values **row-wise** in the input array.


```python
# Input — do not modify
np.random.seed(100)
arr = np.random.randint(1, 11, size=(6, 10))
```

```python
# Write your code here:
def get_row_counts(row):
    # Returns an array tracking value counts from 1 up to 10
    return np.bincount(row, minlength=11)[1:]

row_counts = np.array([get_row_counts(row) for row in arr])
print(row_counts)
```
